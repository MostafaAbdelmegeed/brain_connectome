```python
def investigate_array(arr):
    # Maximum and Minimum
    max_value = np.max(arr)
    min_value = np.min(arr)
    # Mean
    mean_value = np.mean(arr)
    # Variance
    variance_value = np.var(arr)
    # Standard Deviation
    std_deviation = np.std(arr)
    non_zeros = np.count_nonzero(arr)
    ratio = non_zeros / arr.size
    # Print all results
    print(f"Maximum: {max_value}")
    print(f"Minimum: {min_value}")
    print(f"Mean: {mean_value}")
    print(f"Variance: {variance_value}")
    print(f"Standard Deviation: {std_deviation}")
    print(f"Non-Zeros: {non_zeros}")
    print(f"Ratio: {ratio}")
```

```python
base_path = 'C:\\Users\\mosta\OneDrive - UNCG\\Academics\\CSC 699 - Thesis\\data\\ppmi'


def lowercurv(matrix):
    G = nx.from_numpy_array(matrix, parallel_edges= False)
    lrc = LowerORicci(matrix, G)
    K2 = lrc.lower_curvature2()
    return K2

def new_bcurv(matrix):
    G = nx.from_numpy_array(matrix, parallel_edges= False)
    nbc = NewBRicci(matrix, G)
    K3 = nbc.new_whole()
    return K3

# start_time = time.time()
# # Specify the directory path
# directory_path = '/brain/Data2/AD-50'

# # Loop through each file in the directory
# for filename in os.listdir(directory_path):
#     # Construct the full file path
#     file_path = os.path.join(directory_path, filename)
#     # Check if it is a file (and not a directory)
#     if os.path.isfile(file_path):
#         A = np.loadtxt(file_path)
#         W = (A + np.transpose(A))/2
#         K2 = lowercurv(W)
#         K3 = new_bcurv(W)
#         np.savetxt(f'/brain/Data2/Result/AD-50/{filename}_lrc2.txt',K2, fmt='%10.5f', delimiter=' ')
#         np.savetxt(f'/brain/Data2/Result/AD-50/{filename}_new2.txt',K3, fmt='%10.5f', delimiter=' ')
# end_time = time.time()
# execution_time = end_time - start_time
# print(f"Execution time: {execution_time} seconds")

start_time = time.time()

# Walk through all directories and files in the base_path
for root, dirs, files in os.walk(base_path):
    for file in files:
        # Check if 'AAL116_correlation_matrix' is part of the file name
        if 'AAL116_correlation_matrix' in file:
            # Construct full path to file
            open_file_path = os.path.join(root, file)
            lrc_save_file_path = open_file_path.replace('_correlation_matrix.mat', '_lrc_matrix.txt')
            new_save_file_path = open_file_path.replace('_correlation_matrix.mat', '_new_matrix.txt')
            # Load the .mat file
            mat_contents = loadmat(open_file_path)
            A = mat_contents['data']
            W = (A + np.transpose(A))/2
            K2 = lowercurv(W)
            K3 = new_bcurv(W)
            np.savetxt(lrc_save_file_path,K2, fmt='%10.5f', delimiter=' ')
            np.savetxt(new_save_file_path,K3, fmt='%10.5f', delimiter=' ')
            
            
end_time = time.time()
execution_time = end_time - start_time
print(f"Execution time: {execution_time} seconds")
```

```python
control_mat_contents = loadmat(control_example_path)
patient_mat_contents = loadmat(patient_example_path)
```

```python
A_control = control_mat_contents['data']
A_patient = patient_mat_contents['data']
W_control = (A_control + np.transpose(A_control))/2
W_patient = (A_patient + np.transpose(A_patient))/2
```

```python
K_lrc_control = lowercurv(W_control)
K_new_control = new_bcurv(W_control)
K_lrc_patient = lowercurv(W_patient)
K_new_patient = new_bcurv(W_patient)
```

```python
print('---------------------- A_control')
investigate_array(A_control)
print('---------------------- A_patient')
investigate_array(A_patient)
print('---------------------- lrc_control')
investigate_array(K_lrc_control)
print('---------------------- lrc_patient')
investigate_array(K_lrc_patient)
print('---------------------- new_control')
investigate_array(K_new_control)
print('---------------------- new_patient')
investigate_array(K_new_patient)
```
